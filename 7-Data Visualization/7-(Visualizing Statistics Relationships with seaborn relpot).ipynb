```python
sns.relplot(
    data=fmri,
    x='timepoint',
    y='signal',
    kind='line',
    estimator=None
)
```

```python
g = sns.relplot(
    data=fmri,
    x='timepoint',
    y='signal',
    kind='line',
    estimator='max'
)

g.map(
        sns.scatterplot(
        data=fmri,
        x='timepoint',
        y='signal',
        color='red'
    )
)
```

```python
g = sns.relplot(
    data=fmri,
    x='timepoint',
    y='signal',
    kind='line',
    estimator=lambda x: sum(x) / len(x)
)

g.map(
        sns.scatterplot(
        data=fmri,
        x='timepoint',
        y='signal',
        color='red'
    )
)
```
