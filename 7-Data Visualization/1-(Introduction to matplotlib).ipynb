```python
def line_plot(ax, data1, data2, param_dict={}):
    out = ax.plot(data1, data2, **param_dict)
    return out
```

```python
data1, data2, data3, data4 = np.random.randn(4, 100)
fig, axs = plt.subplots(1, 2, figsize=(5, 2.5))
line_plot(axs[0], data1, data2)
```

```python
fig, axs = plt.subplots(1, 2, figsize=(5, 2.5))
line_plot(axs[0], data1, data2, {'marker': 'x'})
```

```python
fig, axs = plt.subplots(1, 2, figsize=(5, 2.5))
line_plot(axs[0], data1, data2, {'marker': '.'})
```

```python
fig, axs = plt.subplots(1, 2, figsize=(5, 2.5))
line_plot(axs[0], data1, data2, {'marker': '.', 'color': 'r'})
```

```python
fig, axs = plt.subplots(1, 2, figsize=(5, 2.5))
line_plot(axs[0], data1, data2, {'marker': '.'})
line_plot(axs[1], data3, data4, {'marker': '.', 'color': 'k'})
```

```python
def scatter_plot(ax, data1, data2, param_dict={}):
    out = ax.scatter(data1, data2, **param_dict)
    return out
```

```python
fig, axs = plt.subplots(1, 2, figsize=(5, 2.5))
scatter_plot(axs[0], data1, data2, {'marker': '.'})
scatter_plot(axs[1], data3, data4, {'marker': 'x', 'color': 'k'})
```
