```python
# Temporary rc settings
with mpl.rc_context({'lines.linewidth': 5, 'lines.linestyle': ':'}):
    plt.plot(data)

plt.plot(data / 3)
```

```python
# Temporary rc settings
with mpl.rc_context({'lines.linewidth': 5, 'lines.linestyle': ':'}):
    mpl.rcdefaults()
    plt.plot(data)

plt.plot(data / 3)
```

```python
# Temporary rc settings

mpl.rcdefaults()

with mpl.rc_context({'lines.linewidth': 5, 'lines.linestyle': ':'}):
    plt.plot(data)

plt.plot(data / 3)
```

```python
@mpl.rc_context({'lines.linestyle': '-', 'lines.linewidth': 0.5})
def plot_data():
    plt.plot(data)

plot_data()
```

```python
@mpl.rc_context({'lines.linestyle': '--', 'lines.linewidth': 3})
def plot_data():
    plt.plot(data)

plot_data()
```

```python
@mpl.rc_context({})
def plot_data():
    plt.plot(data)

plot_data()
```
