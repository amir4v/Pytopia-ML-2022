```python
class DataVis:
    def __init__(self, x, y):
        self._x = x
        self._y = y
    
    def plot(self, ax=None, **kwargs):
        if ax is None:
            ax = plt.gca()
        ax.plot(self._x, self._y, **kwargs)
        ax.set_title('Yeah')

d = DataVis([1,3,5,6], [7,1,9,3])
d.plot()
```

```python
class DataVis:
    def __init__(self, x, y):
        self._x = x
        self._y = y
    
    def plot(self, ax=None, **kwargs):
        if ax is None:
            ax = plt.gca()
        ax.scatter(self._x, self._y, **kwargs)
        ax.set_title('Yeah')

d = DataVis([1,3,5,6], [7,1,9,3])
d.plot(color='red')
```
