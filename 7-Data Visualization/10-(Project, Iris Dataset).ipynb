```python
import seaborn as sns

%matplotlib inline
```

```python
custom_rc_params = {
    'figure.figsize': (6,4),
}

sns.set_theme(
    style='whitegrid',
    font_scale=1.2,
    context='notebook', # paper, notebook, poster, talk
    rc=custom_rc_params,
)
```

```python
iris = sns.load_dataset('iris')
```

```python
iris.head()
```

```python
df = iris
```

```python
df.sample(3)
```

```python
sns.histplot(
    data=df,
    x='sepal_length'
)
```

```python
sns.displot(
    data=df,
    x='sepal_length',
    aspect=2
)
```
